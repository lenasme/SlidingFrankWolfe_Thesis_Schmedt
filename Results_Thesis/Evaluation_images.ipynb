```python
seed = 13
grid_size= 100
deltas = [0.08-0.005, 0.09+0.005]
max_jumps = 3 
variance= 0.01
reg_param = 1/0.028 * np.sqrt(variance)

import pickle
import matplotlib.pyplot as plt


fw_objective_co5 = "objective_fw_cutoff5_seed13.pkl"
sfw_objective_co5 = "objective_sfw_cutoff5_seed13.pkl"
fw_objective_co18 = "objective_fw_cutoff18_seed13.pkl"
sfw_objective_co18 = "objective_sfw_cutoff18_seed13.pkl"


with open(fw_objective_co5, "rb") as f:
    fw_objectives_co5 = pickle.load(f)

with open(sfw_objective_co5, "rb") as f:
    sfw_objectives_co5 = pickle.load(f)
    
with open(fw_objective_co18, "rb") as f:
    fw_objectives_co18 = pickle.load(f)

with open(sfw_objective_co18, "rb") as f:
    sfw_objectives_co18 = pickle.load(f)


plt.figure(figsize=(10, 6))
opt = compute_objective_ground_truth(grid_size, max_jumps, seed, deltas, reg_param)
plt.axhline(opt, linestyle="--", color="black", label="ground truth objective", linewidth=2)
plt.plot(fw_objectives_co5, label="classic Frank-Wolfe,  $\Phi$ = 5", linewidth=2)
plt.plot(sfw_objectives_co5, label="Sliding Frank-Wolfe,  $\Phi$= 5", linewidth=2)
plt.plot(fw_objectives_co18, label="classic Frank-Wolfe,  $\Phi$ = 18", linewidth=2)
plt.plot(sfw_objectives_co18, label="Sliding Frank-Wolfe,  $\Phi$ = 18", linewidth=2)

plt.xlabel("Iteration")
plt.ylabel(r"$J_{\alpha}^{L^{2,0}}$")
plt.yscale("log")

plt.legend(fontsize = 16)
plt.grid(True)
plt.tight_layout()
plt.savefig(fr"C:\Lena\Universität\Inhaltlich\Master\AMasterarbeit\Masterarbeit_Dokument\objective_graph_seed{seed}.png", dpi=300)
plt.show()
```
